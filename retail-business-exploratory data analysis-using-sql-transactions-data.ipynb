```python
# Section 4: Data Preparation and Processing

# Activity 1: Inspect the first few rows of a dataframe called `veg_txn_df` to gain an overview of its columns and values.
veg_txn_df.head()

# Activity 2: Conduct an Exploratory Data Analysis (EDA) on a dataframe to examine the data types and column names. 
veg_txn_df.dtypes

# Activity 3: Create new_column_names dataframe with the desired column names on the table
new_column_names = ['txn_date', 'txn_time', 'item_code', 'qty_sold(kg)', 'unit_selling_px_rmb/kg','sale/return', 'discount(%)']

# Activity 4: rename the columns in place
veg_txn_df.rename(columns=dict(zip(veg_txn_df.columns, new_column_names)), inplace=True)

# Activity 5: convert the date column to datetime datatype
veg_txn_df['txn_date'] = pd.to_datetime(veg_txn_df['txn_date'])

# Activity 6: add a new column with the formatted date
veg_txn_df['day_of_week'] = veg_txn_df['txn_date'].dt.strftime('%A')

# Activity 7: Data Type and Column Name Cleaning
veg_txn_df = veg_txn_df.astype({'txn_time': 'timedelta64[s]','discount(%)': 'bool'})
veg_txn_df['txn_date'] = pd.to_datetime(veg_txn_df['txn_date'])
veg_txn_df['txn_date'] = veg_txn_df['txn_date'].dt.floor('D')
veg_txn_df['txn_time'] = veg_txn_df['txn_time'].dt.total_seconds()
veg_txn_df['txn_time'] = pd.to_datetime(veg_txn_df['txn_time'],unit='s').dt.strftime('%H:%M:%S')
```

# Query 2: Top 5 Highest Selling Item_codes by Quantity_sold and Date
**Summary of queries:**
The queries aim to identify the top 5 highest selling items and their corresponding transaction dates based on the quantity sold in kilograms. They retrieve data from the "veg_txn_df" table, which appears to contain information about transactions related to the sale of vegetables. The queries select specific columns, format the date, and sort the data in descending order based on the quantity sold, ultimately limiting the results to the top 5 items.

**Rationale:**
The rationale behind these queries is to gain insights into the top-performing items within the dataset. By identifying the items that have the highest quantity sold and their corresponding transaction dates, businesses and analysts can make informed decisions regarding inventory management, pricing strategies, customer preferences, and overall sales performance.

**Relevance:**
These queries are highly relevant in the context of businesses that deal with the sale of vegetables or similar products. Knowing which items are the highest sellers and when these sales occur can inform a range of business strategies, including stock management, pricing adjustments, marketing efforts, and product development. This information allows businesses to respond effectively to market demands and optimize their operations.

**Necessity:**
The necessity of these queries is underscored by the need for data-driven decision-making in business operations. Understanding the top-selling items and their sales dates is crucial for businesses to remain competitive, efficient, and customer-focused. It enables businesses to maximize profits, minimize wastage, and cater to consumer preferences, ultimately contributing to the success and sustainability of the enterprise.

```python
# Query 2: Define SQL Query to Identify the Top 5 Highest Selling Item_codes by Quantity_sold and Date
query =text(
'''

select STRFTIME('%Y-%m-%d',[txn_date]) as 'date', item_code, [unit_selling_px_rmb/kg], [qty_sold(kg)]
from veg_txn_df
Order by [qty_sold(kg)] DESC
limit 5;

'''

)

# Execute the SQL Query and Retrieve the Results into a DataFrame
df_sql = pd.read_sql_query(query, con)

# Display the Top 5 Highest Selling items and dates by Quantity Sold
df_sql.head()

# Below is the query to visualize the above sql query

# Data for the chart
dates = df_sql['date']
items = df_sql['item_code']
quantities = df_sql['qty_sold(kg)']

# Set up the chart
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Top 5 Highest Selling Item_codes by Quantity_sold and Date')
ax.set_xlabel('Date')
ax.set_ylabel('Quantity Sold (kg)')

# Plot the data
colors = ['blue', 'green', 'red', 'orange', 'purple']
for i in range(len(items)):
    ax.bar(dates[i], quantities[i], color=colors[i], label=items[i])

# Add the legend
ax.legend()

# Show the chart
plt.show()
```

# Query 3: Top 5 Lowest Selling Item_codes by Quantity_sold and Date
This query is part of the exploratory data analysis (EDA) for the 'veg_txn_df' dataset. It aims to identify the top 5 items with the lowest quantity sold in kilograms, providing insights into the least popular items in terms of sales volume.

```python
# Query 3: Define SQL Query to Lowest Selling Item_code Quantity_sold and Date
query =text(
'''

select STRFTIME('%Y-%m-%d',[txn_date]) as 'Date', item_code as 'Item_Code', [unit_selling_px_rmb/kg] AS 'Unit_Price(RMB)/Kg', [qty_sold(kg)]
from veg_txn_df
where [qty_sold(kg)] >=0  --To exclude returned items
Order by [qty_sold(kg)] ASC
limit 5;
'''

)

# Execute the SQL Query and Retrieve the Results into a DataFrame
df_sql = pd.read_sql_query(query, con)

# Display the Items with the Lowest Quantity Sold
df_sql.head()

# Let's plot the query.
# Assuming you already have df_sql DataFrame from your query
# It should contain columns: 'item_code' and 'qty_sold(kg)'

# Sort the DataFrame by quantity sold in ascending order
df_sql_sorted = df_sql.sort_values(by='qty_sold(kg)').head(5)

# Create a list of item codes and their corresponding quantities sold
item_codes = df_sql_sorted['Item_Code'].tolist()
quantities_sold = df_sql_sorted['qty_sold(kg)'].tolist()

# Create a radar chart
categories = item_codes
values = quantities_sold

# Number of variables
num_vars = len(categories)

# Calculate angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# Ensure the values list has the same number of elements as angles
values += values[:1]

# Plot the radar chart
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
plt.xticks(angles[:-1], categories, color='grey', size=10)

ax.set_rlabel_position(0)
plt.yticks([min(quantities_sold), max(quantities_sold)], ['Min', 'Max'], color='grey', size=8)
plt.ylim(min(quantities_sold) - 5, max(quantities_sold) + 5)

ax.fill(angles, values, 'b', alpha=0.1)
plt.title('Top 5 Lowest Selling Items - Quantity Sold Radar Chart', size=20, color='b')

plt.show()
```

# Query 10: Total Kilograms Sold Annually
- **Rationale**: Calculates and presents the total kilograms sold annually.
- **Relevance**: Offers insights into annual sales patterns.
- **Necessity**: Helps in annual performance assessment.

```python
# Query 10: Define SQL Query for Total Kilograms Sold Annually
query = text('''
SELECT strftime('%Y', [txn_date]) as "year", sum([qty_sold(kg)]) as "kg_sold/period"
FROM veg_txn_df
GROUP BY "year"
''')

# Execute the SQL Query and Retrieve the Results into a DataFrame
df_sql = pd.read_sql_query(query, con)

# Display the Total Kilograms Sold Annually
df_sql.head()

# Data visualization below

# SQL query result
df_sql = pd.read_sql_query(query, con)

# Create a bar chart with custom colors and annotations
plt.figure(figsize=(12, 8))
colors = plt.cm.Paired(range(len(df_sql)))
bars = plt.bar(df_sql['year'], df_sql['kg_sold/period'], color=colors)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Total Kilograms Sold')
plt.title('Total Kilograms Sold Annually')

# Annotate the data points with kg_sold/period values
for bar, value in zip(bars, df_sql['kg_sold/period']):
    plt.text(bar.get_x() + bar.get_width() / 2, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Show the plot
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# Query 44: Total Qty Sold by Week of the Month
This query aims to extract information from the `veg_txn_df` dataset related to the year, month, and quantity sold for each transaction. Here's a breakdown of the query. The query essentially calculates the total quantity sold (in kilograms) for each year, month, and week of the month in the `veg_txn_df` dataset. The results will be grouped accordingly and ordered by the 'txn_date' column.

## Necessity of the Query:
- **Sales Pattern Analysis:** Understanding how sales vary during different weeks of the month can provide insights into customer buying behavior.
- **Inventory Management:** Valuable for stock replenishment planning to meet varying demand throughout the month.
- **Sales Forecasting:** Historical week-wise sales data helps in forecasting future sales more accurately.
- **Marketing Strategy:** Adjusting marketing efforts according to sales patterns can be beneficial for promotion and pricing strategies.

```python
# Query 44: Define SQL Query for Total Qty Sold by Week of the Month

query =text(
'''
select
  strftime('%Y', [txn_date]) as 'yr',
  CASE 
   WHEN strftime('%m', [txn_date]) <= '01' THEN 'Jan '
   WHEN strftime('%m', [txn_date]) <= '02' THEN 'Feb '
   WHEN strftime('%m', [txn_date]) <= '03' THEN 'Mar '
   WHEN strftime('%m', [txn_date]) <= '04' THEN 'Apr '
   WHEN strftime('%m', [txn_date]) <= '05' THEN 'May '
   WHEN strftime('%m', [txn_date]) <= '06' THEN 'Jun '
   WHEN strftime('%m', [txn_date]) <= '07' THEN 'Jul '
   WHEN strftime('%m', [txn_date]) <= '08' THEN 'Aug '
   WHEN strftime('%m', [txn_date]) <= '09' THEN 'Sep '
   WHEN strftime('%m', [txn_date]) <= '10' THEN 'Oct '
   WHEN strftime('%m', [txn_date]) <= '11' THEN 'Nov '
   WHEN strftime('%m', [txn_date]) <= '12' THEN 'Dec '
  END AS monthly_period,
  CASE 
   WHEN strftime('%d', [txn_date]) BETWEEN '01' AND '07' THEN 'Week 1 ' 
   WHEN strftime('%d', [txn_date]) BETWEEN '08' AND '14' THEN 'Week 2 '
   WHEN strftime('%d', [txn_date]) BETWEEN '15' AND '21' THEN 'Week 3 '
   WHEN strftime('%d', [txn_date]) BETWEEN '22' AND '28' THEN 'Week 4 '
   ELSE 'Week 5 '
  END AS week_of_month,
  SUM([qty_sold(kg)]) as total_kgs_sold
FROM
  veg_txn_df
Group by 'yr', monthly_period, week_of_month
ORDER BY [txn_date];

'''

)

df_sql = pd.read_sql_query(query,con)

df_sql.head(1000)
```

# Query 45: Top 10 Weeks with Highest Total Qty Sold by Week of the Month
This query aims to identify and list the top 10 weeks with the highest total quantity sold by week of the month for the years 2020, 2021, 2022, and 2023.

## Necessity of the Query:
- **Sales Performance Evaluation:** Identifying peak sales weeks for each month helps in evaluating sales performance.
- **Demand Forecasting:** This analysis can be valuable for forecasting demand in future weeks.
- **Marketing Focus:** Knowing which weeks are most significant for sales allows businesses to target their marketing efforts more effectively.
- **Inventory Management:** Helps in managing stock levels during high-demand weeks.

```python
# Query 45: Define SQL Query to Find Top 10 Weeks with Highest Total Qty Sold by Week of the Month

query =text(
'''
WITH cte as (
SELECT 
  CASE
   WHEN strftime('%Y', [txn_date]) = '2020' THEN 'Yr_2020'
   WHEN strftime('%Y', [txn_date]) = '2021' THEN 'Yr_2021'
   WHEN strftime('%Y', [txn_date]) = '2022' THEN 'Yr_2022'
   WHEN strftime('%Y', [txn_date]) = '2023' THEN 'Yr_2023'
   ELSE 'Yr_NA'
  END AS year,
  CASE 
   WHEN strftime('%m', [txn_date]) = '01' THEN 'Jan '  
   WHEN strftime('%m', [txn_date]) = '02' THEN 'Feb '
   WHEN strftime('%m', [txn_date]) = '03' THEN 'Mar '
   WHEN strftime('%m', [txn_date]) = '04' THEN 'Apr '
   WHEN strftime('%m', [txn_date]) = '05' THEN 'May '
   WHEN strftime('%m', [txn_date]) = '06' THEN 'Jun '
   WHEN strftime('%m', [txn_date]) = '07' THEN 'Jul '
   WHEN strftime('%m', [txn_date]) = '08' THEN 'Aug '
   WHEN strftime('%m', [txn_date]) = '09' THEN 'Sep '
   WHEN strftime('%m', [txn_date]) = '10' THEN 'Oct '
   WHEN strftime('%m', [txn_date]) = '11' THEN 'Nov '
   WHEN strftime('%m', [txn_date]) = '12' THEN 'Dec '
  END AS monthly_period,
  CASE 
   WHEN strftime('%d', [txn_date]) BETWEEN '01' AND '07' THEN 'Week 1 ' 
   WHEN strftime('%d', [txn_date]) BETWEEN '08' AND '14' THEN 'Week 2 '
   WHEN strftime('%d', [txn_date]) BETWEEN '15' AND '21' THEN 'Week 3 '
   WHEN strftime('%d', [txn_date]) BETWEEN '22' AND '28' THEN 'Week 4 '
   ELSE 'Week 5 '
  END AS week_of_month,      
  SUM([qty_sold(kg)]) as total_kgs_sold
FROM veg_txn_df  
GROUP BY 'yr', monthly_period, week_of_month 
)

SELECT DISTINCT year, monthly_period, week_of_month, total_kgs_sold  
FROM cte
ORDER BY total_kgs_sold DESC
LIMIT 10;

'''

)

df_sql = pd.read_sql_query(query,con)

df_sql.head(1000)
```
